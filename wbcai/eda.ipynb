```python

#lists
image_data_dict = {k:[] for k in image_meta_tags}
lens = []
labels1 = []
labels2 = []
coords = []
fnames = []
#loop
for f in filenames:
    fnames.append(f)
    f = open(DATA_DIR + '/jsons/' + f + '.json')
    data = json.load(f)
    lab1 = []
    lab2 = []
    coord = []
    #collect data about LABELS
    number_cells = int(data['Cell Numbers'])
    for i in range(0,number_cells):
        cell_data = data['Cell_'+str(i)]
        lab1.append(cell_data['Label1'])
        lab2.append(cell_data['Label2'])
        x1=str(cell_data['x1'])
        x2=str(cell_data['x2'])
        y1=str(cell_data['y1'])
        y2=str(cell_data['y2'])
        coord.append(','.join([x1,x2,y1,y2]))
#new_img=img[y1:y2,x1:x2,:]
    coord = '|'.join(coord)
    coords.append(coord)
    
    lab1 = ['None' if v is None else v for v in lab1]
    labels1.append(','.join(lab1))
    lab2 = ['None' if v is None else v for v in lab2]
    #print(lab2)
    labels2.append(','.join(lab2))
    #collect data about IMAGES
    for i in image_meta_tags:
        image_data_dict[i].append(data[i])
```

```python
blooddf = pd.DataFrame(image_data_dict)
```

```python
def flatten_stringlist(LIST):
    # ['one', 'five', 'one,five', 'six, seven, one']
    # flattens lists like 
    newl = []
    for i in LIST:
        if i.count(',') > 0:
            multiple_labels = i.split(',')
            for x in multiple_labels:
                newl.append(x)
        else:
            newl.append(i)
    return newl
```

```python
unique_labels = list(set(flatten_stringlist(blooddf.labels1)).union(set(flatten_stringlist(blooddf.labels2))))
all_labels = flatten_stringlist(blooddf.labels1) + flatten_stringlist(blooddf.labels2)
count_dict = {}
for i in unique_labels:
    count_dict[i] = all_labels.count(i)
count_dict
```

```python
label1counts = []
label2counts = []
for i in count_dict.keys():
    label1counts.append(' '.join(list(blooddf.labels1)).count(i))
    label2counts.append(' '.join(list(blooddf.labels2)).count(i))
    
count_df1 = pd.DataFrame({'type':count_dict.keys(),'counts':label1counts, 'labeler':['labeler 1'] * len(count_dict.keys())})
count_df2 = pd.DataFrame({'type':count_dict.keys(),'counts':label2counts, 'labeler':['labeler 2'] * len(count_dict.keys())})


label_df = pd.concat([pd.DataFrame(count_df1),pd.DataFrame(count_df2)])

ax = sns.barplot(x='type',y='counts', hue='labeler', data=label_df)
plt.xticks(rotation=90)
plt.show()
```

```python

import cv2
import numpy as np

def crop_micro(pick,new_dir=''):
    # load image as grayscale
    img = plot_single_label(pick)

    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # threshold 
    thresh = cv2.threshold(gray, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)[1]
    hh, ww = thresh.shape

    # make bottom 2 rows black where they are white the full width of the image
    thresh[hh-3:hh, 0:ww] = 0

    # get bounds of white pixels
    white = np.where(thresh==255)
    xmin, ymin, xmax, ymax = np.min(white[1]), np.min(white[0]), np.max(white[1]), np.max(white[0])
    print(xmin,xmax,ymin,ymax)

    # crop the image at the bounds adding back the two blackened rows at the bottom
    crop = img[ymin:ymax+3, xmin:xmax]
    return crop, (xmin,xmax,ymin,ymax)
```
